```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

# take this as REFERENCE; you'll have to manually input your final values before running
# always tailor your analysis to the data collected

nearest = list(map(lambda x: x * 3.28084, [2.486, 2.412, 2.501, 2.313, 2.439]))
middle = list(map(lambda x: x * 3.28084, [1.673, 1.829, 2.364, 2.390, 2.056]))
farthest = list(map(lambda x: x * 3.28084, [2.247, 2.208, 2.168, 1.891, 1.992]))
set = [farthest, middle, nearest]

nearstd = np.std(nearest)
midstd = np.std(middle)
farstd = np.std(farthest)

nearmed = np.median(nearest)
midmed = np.median(middle)
farmed = np.median(farthest)

print("STDEVs: (Farthest: %.3f, Middle: %.3f, Nearest: %.3f)" % (farstd, midstd, nearstd))
print("MEDIANs: (Farthest: %.3f, Middle: %.3f, Nearest: %.3f)" % (farmed, midmed, nearmed))

fig = plt.figure(figsize = (10, 7))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_ylabel("Turning Radius (feet)")
ax.set_xlabel("Sway Bar Setting")
bp = ax.boxplot(set, labels=["Farthest (3)" + "\n" + "n=5", "Middle (2)" + "\n" + "n=5", "Nearest (1)" + "\n" + "n=5"])
plt.title("TR24 Turning Radius vs. Sway Bar Setting")
plt.show()

fstat, p = sp.stats.f_oneway(farthest, middle, nearest)
print(f"p-value is {p}")
```
